```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("telecom_customer_churn.csv")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Streaming TV','Streaming Music' ,'Streaming Movies', 'Device Protection Plan','Churn Reason','Online Security', 'Online Backup', 'Device Protection Plan','Total Refunds','Customer ID', 'Gender', 'Age', 'Married', 'Number of Dependents',], axis=1 )
```

```python
df.columns
```

```python
df = df.drop(['Premium Tech Support', 'Unlimited Data','Total Extra Data Charges','Churn Category','Paperless Billing','Total Revenue','City', 'Zip Code','Tenure in Months'], axis=1 )
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df['Avg Monthly Long Distance Charges'].fillna(df['Avg Monthly Long Distance Charges'].mean(), inplace =True)
```

```python
df['Avg Monthly GB Download'].fillna(df['Avg Monthly GB Download'].mean(), inplace =True)
```

```python
df['Internet Type'].fillna('no internet type', inplace =True)
```

```python
df.isnull().sum()
```

```python
df['Internet Type'].replace(['Fiber Optic', 'no internet type', 'Cable','DSL'], [1,0,2,3], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Offer']=df['Offer'].fillna(method='ffill')
```

```python
df['Multiple Lines']=df['Multiple Lines'].fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Offer'])
```

```python
df.head()
```

```python
df['Internet Service'].replace(['Yes', 'No'],[1,0], inplace=True)
df['Multiple Lines'].replace(['Yes', 'No'],[1,0], inplace=True)
df['Phone Service'].replace(['Yes', 'No'],[1,0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder= LabelEncoder()
```

```python
df['Offer']= label_encoder.fit_transform(df['Offer'])
```

```python
df['Contract']= label_encoder.fit_transform(df['Contract'])
```

```python
df.head()
```

```python
df['Payment Method']= label_encoder.fit_transform(df['Payment Method'])
```

```python
df['Customer Status']= label_encoder.fit_transform(df['Customer Status'])
```
